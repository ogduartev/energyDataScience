Lectura del archivo de datos:

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

target_url="https://raw.githubusercontent.com/ogduartev/energyDataScience/main/data/user/sample01.csv";
df = pd.read_csv(target_url);
df
```

Eliminación de las columnas inútiles

```python
cols_null=["id_medidor *","potencia_reactiva_kvar","thd_tension","thd_corriente","Potencia aparente (kva)","Frecuencia (Hz)","Generación de energía (Kw/h)"]
df=df.drop(columns=cols_null);
```

Convertir las columnas numéricas

```python
cols_num=["energia_kwh","potencia_kw",
          "tension_fase_a","tension_fase_b","tension_fase_c",
          "corriente_fase_a","corriente_fase_b","corriente_fase_c",
          "factor_potencia",
          "potencia_fase_a","potencia_fase_b","potencia_fase_c"]
for col in cols_num:
    df[col] = df[col].str.replace(',','.') 
    df[col] = df[col].astype(float)
col="Energia reactiva (kvar)"
df[col] = df[col].astype(float)
```

Interpretar el formato de fecha

```python
from datetime import datetime

df['fecha']=pd.to_datetime(df['fecha'],format='%d/%m/%Y %H:%M:%S')
```

Ordenamiento por fecha

```python
df.sort_values('fecha')
```

Lecturas de potencias por fase erróneas

```python
limite_pot=1000.0
for ind in df.index:
    if (df['potencia_fase_a'][ind] > limite_pot):
        df['potencia_fase_a'][ind]=0.0
    if (df['potencia_fase_b'][ind] > limite_pot):
        df['potencia_fase_b'][ind]=0.0
    if (df['potencia_fase_c'][ind] > limite_pot):
        df['potencia_fase_c'][ind]=0.0
    df['potencia_kw'][ind]=df['potencia_fase_a'][ind]+df['potencia_fase_b'][ind]+df['potencia_fase_c'][ind]
```

```python
df.describe()
```

```python
dia=[''] * df["fecha"].size
hora=[''] * df["fecha"].size
for ind in df.index:
  x=df["fecha"][ind]
  dia[ind]=x.strftime("%Y-%m-%d")
  hora[ind]=x.strftime("%H")
df["dia"]=dia
df["hora"]=hora
dd=df.groupby(['dia','hora'])[['potencia_kw']].mean().unstack()
```
