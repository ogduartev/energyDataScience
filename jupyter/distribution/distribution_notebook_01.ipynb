```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

target_url="https://raw.githubusercontent.com/ogduartev/energyDataScience/main/data/distribution/sample01.csv"
df = pd.read_csv(target_url,sep=';',encoding='iso8859-1',low_memory=False)
```

```python
dd=df.groupby(['COD_MEDIDOR','HORA'])[['CONSUMO']].mean().unstack()
```

```python
dd.head()
```

```python
import skfuzzy as fuzz

data=dd['CONSUMO']

conjuntos=5;
m=2.0;
error=1e-9;
max_iter=20;
cntr, u, u0, d, jm, p, fpc=fuzz.cmeans(data, conjuntos, m, error, max_iter);
print("centros:");
print(cntr);
```
